```python
import numpy as np
import pandas as pd
import seaborn as sns
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
```

# Exploratory Data Analysis

```python
wm = pd.read_csv("data/waste_management.csv", index_col = 0)
```

```python
wm.info()
```

```python
# NAs across the dataset
wm.apply(lambda x: sum(x.isnull().values), axis = 0)
```

```python
# Rows containing NAs
sum(wm.apply(lambda x: sum(x.isnull().values), axis = 1)>0)
```

```python
wm['Abfaelle pro Bevoelkerung'] = np.where(wm['Bevoelkerung']!=0, wm['Abfaelle_gesamt']/wm['Bevoelkerung'], 0)
```

```python
# Remove rows with missing values
wm_df = wm.dropna()
```

```python
waste_by_region_bxplt = sns.boxplot(
    x = 'Region',
    y = 'Abfaelle_gesamt',
    data = wm_df,
    showfliers = False
)

waste_by_region_bxplt = waste_by_region_bxplt.set_xticklabels(
    labels = waste_by_region_bxplt.get_xticklabels(),
    rotation = 90
)
```

```python
waste_by_geological_location = sns.boxplot(
    x = 'Geologischer_Indikator',
    y = 'Abfaelle_gesamt',
    data = wm_df,
    showfliers = False
)

waste_by_geological_location = waste_by_geological_location.set_xticklabels(['South', 'Mid', 'North'])
```

```python
waste_by_pop = sns.scatterplot(
    x = "Bevoelkerung",
    y = "Abfaelle_gesamt",
    hue = "Geologischer_Indikator",
    style = "Urbanisierungsgrad",
    data = wm_df
)
```

```python
wm_df_nums = wm_df.select_dtypes(include=np.number)
wm_df_corr= wm_df_nums.dropna().corr()
wm_corrplot = sns.heatmap(
    wm_df_corr,
    annot = False,
)
```
